Q1. What is the difference between descriptive statistics and inferential
statistics? Explain with examples.

  - Descriptive Statistics

1) Definition: Descriptive statistics deals with collecting, summarizing, and presenting data in a meaningful way.

2) Purpose: It describes what the data shows but does not make predictions or generalizations beyond the dataset.

3) Examples of descriptive statistics tools:
    
    - Measures of central tendency: Mean, Median, Mode
    - Measures of dispersion: Range, Variance, Standard Deviation
    - Data visualization: Graphs, Charts, Tables

Example:

  A teacher calculates the average marks of 50 students in her class.
  - Average = 72, Standard deviation = 10.

This tells us how students performed in that class only. It doesn’t predict about students in other schools.

  - Inferential Statistics

1) Definition: Inferential statistics uses sample data to make predictions, generalizations, or decisions about a larger population.

2) Purpose: It goes beyond the data to make inferences about the population, often using probability theory.

3) Examples of inferential statistics tools:

    - Hypothesis testing (t-test, chi-square test)
    - Confidence intervals
    - Regression analysis
    - ANOVA

Example:

  A researcher surveys 200 people from a city about their favorite soft drink.
  - From the sample, 60% prefer Brand A.
  - Using inferential statistics, the researcher predicts that about 60% of the entire city’s population prefers Brand A (with some margin of error).


Key Difference (Simple)

- Descriptive: “What does the data say?” (Summarizes existing data)
- Inferential: “What can we conclude or predict?” (Generalizes from sample to population)



Q2. What is sampling in statistics? Explain the differences between random
and stratified sampling.

- What is Sampling in Statistics?

In statistics, sampling is the process of selecting a subset (sample) from a larger group (population) so that we can study the sample and make conclusions about the population.

    - Population → the entire group we want information about.
    - Sample → a smaller, manageable part of the population.

Sampling is used because studying the whole population is often expensive, time-consuming, or impossible.


Random Sampling :
  - In random sampling, every member of the population has an equal chance of being selected.
  - It is unbiased and simple to understand.
  - Example: If you have 1,000 students in a college, and you randomly pick 100 students using a lottery or computer-generated random numbers, that’s random sampling.

  Pros:
      - Minimizes selection bias.
      - Easy to apply.
  
  Cons:
      - Doesn’t guarantee representation of all subgroups (e.g., gender, age groups).

Stratified Sampling :
  - In stratified sampling, the population is divided into strata (subgroups) based on certain characteristics (e.g., gender, income level, education).
  - Then, a random sample is taken from each subgroup in proportion to its size in the population.
  - Example: If a college has 60% male and 40% female students, and you want a sample of 100 students, you would randomly select 60 males and 40 females → stratified sampling.

  Pros:
      - Ensures all important subgroups are represented.
      - Provides more accurate and reliable results than simple random sampling.

  Cons:
      - Requires more information about the population.
      - More complex to implement.


In short:
  - Random sampling = equal chance, simple, but may not represent all subgroups.
  - Stratified sampling = divides population into subgroups, then samples, ensuring better representation.
  




Q3. Define mean, median, and mode. Explain why these measures of central
tendency are important.

- Mean (Arithmetic Average):
  
  The mean is the sum of all values in a dataset divided by the number of values.

  Mean = Sum of all observations​ / Number of observations

  - Example: For data [2, 4, 6, 8],

Mean = 2 + 4 + 6 + 8 / 4
       
       = 20 / 4
      
       = 5

- Median (Middle Value):

  The median is the middle value when the data is arranged in ascending (or descending) order.

  - If the number of values is odd, it is the middle value.
  - If the number of values is even, it is the average of the two middle values.

  Example: For data [2, 4, 6, 8, 10], the median is 6 (middle value).

  For [2, 4, 6, 8], the median is 4+6/2 10/2 = 5


- Mode (Most Frequent Value):

  The mode is the value that occurs most frequently in the dataset.
    - A dataset may have no mode, one mode (unimodal), or multiple modes (bimodal/multimodal).
  
  Example: For data [2, 4, 4, 6, 8, 8, 8], the mode is 8 (appears most often).


Why These Measures Are Important (Central Tendency):
- Mean: Represents the overall average, useful for data with no extreme outliers.
- Median: Represents the middle value, less affected by outliers or skewed data.
- Mode: Shows the most common value, useful in categorical or frequency-based data.

Together, they summarize a dataset with a single value that represents the “center” of the data distribution, making it easier to understand and compare datasets.

Q4. Explain skewness and kurtosis. What does a positive skew imply about the
data?

Skewness
  - Definition: Skewness measures the asymmetry of a data distribution.
  - If the data is perfectly symmetric (like a normal distribution), skewness = 0.
  - Types:
    - Positive Skew (Right-skewed): Tail is stretched more towards the right (higher values).
    - Negative Skew (Left-skewed): Tail is stretched more towards the left (lower values).

  What a Positive Skew Implies:
    - Most of the data values are concentrated on the left, with a long tail to the right.
    - Mean > Median > Mode.
    - Example: Income distribution (a few very high incomes pull the mean to the right).

Kurtosis
  - Definition: Kurtosis measures the "tailedness" or peakedness of the distribution compared to a normal distribution.
  - Types:
    - Mesokurtic (kurtosis ≈ 3): Normal distribution.
    - Leptokurtic (kurtosis > 3): Distribution is more peaked with heavier tails (more outliers).
    - Platykurtic (kurtosis < 3): Distribution is flatter with lighter tails (fewer outliers).

In short:
  - Skewness → tells about asymmetry.
  - Kurtosis → tells about tail thickness/peakedness
  - Positive skew → long right tail, mean pulled right.
  


Q5. : Implement a Python program to compute the mean, median, and mode of
a given list of numbers.

      numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


- Here’s a simple Python program to calculate the mean, median, and mode for the given list of numbers:




```python
# Example Code:

import statistics as stats

# Given list
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Mean
mean_val = stats.mean(numbers)

# Median
median_val = stats.median(numbers)

# Mode
mode_val = stats.mode(numbers)

print("Numbers:", numbers)
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

Q6. Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
            
            list_x = [10, 20, 30, 40, 50]
            list_y = [15, 25, 35, 45, 60]





```python
# Code:

import numpy as np

# Given data
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert to numpy arrays
x = np.array(list_x)
y = np.array(list_y)

# Covariance matrix
cov_matrix = np.cov(x, y, ddof=0)   # population covariance
covariance = cov_matrix[0, 1]

# Correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

print("Covariance:", covariance)
print("Correlation coefficient:", correlation)
```

Q7. Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
  
  data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


  

```python
# Example Code:

import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Draw boxplot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Boxplot of Data")
plt.xlabel("Values")
plt.show()

# --- Outlier Detection using IQR Rule ---
Q1 = np.percentile(data, 25)   # First quartile
Q3 = np.percentile(data, 75)   # Third quartile
IQR = Q3 - Q1                  # Interquartile range

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}")
print(f"Lower Bound = {lower_bound}, Upper Bound = {upper_bound}")
print("Outliers:", outliers)
```

Q8. You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and daily sales.
- Explain how you would use covariance and correlation to explore this
relationship.
- Write Python code to compute the correlation between the two lists:

    - advertising_spend = [200, 250, 300, 400, 500]

      daily_sales = [2200, 2450, 2750, 3200, 4000]



Explanation

1) Covariance
  - Covariance measures how two variables vary together.
  - If both variables increase together, the covariance is positive.
  - If one increases while the other decreases, the covariance is negative.
  - However, covariance is scale-dependent (its value changes with units), so it only tells us the direction of the relationship, not its strength.

2) Correlation
  - Correlation is the standardized form of covariance.
  - It measures both direction and strength of the relationship.
  - It always lies between -1 and +1:
    - +1 → Perfect positive relationship (both move together)
    - -1 → Perfect negative relationship (move in opposite directions)
    - 0 → No linear relationship

In this case, correlation is more useful because it gives a clear picture of how strongly advertising spend is related to daily sales.


```python
# Code:

import numpy as np

# Data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute covariance matrix
cov_matrix = np.cov(advertising_spend, daily_sales)
print("Covariance Matrix:\n", cov_matrix)

# Compute correlation coefficient
correlation = np.corrcoef(advertising_spend, daily_sales)
print("Correlation Coefficient:\n", correlation)
```

Interpretation
  - The correlation ≈ 0.99, which is very close to +1.
  - This means there is a very strong positive relationship: as advertising spend increases, daily sales also increase.
  

Q9. Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
- Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
- Write Python code to create a histogram using Matplotlib for the survey data:
  - survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]


  


Summary Statistics & Visualizations

 To understand the distribution of customer satisfaction scores (scale 1–10), we should look at:

    Summary Statistics
    - Mean (average): Shows the central tendency.
    - Median: Middle score, less affected by outliers.
    - Mode: Most frequently given score.
    - Standard Deviation: Tells us how spread out the scores are.
    - Range (min & max): Smallest and largest scores.

    Visualizations
    - Histogram: Shows frequency distribution of scores.
    - Boxplot: Useful to detect spread and outliers.
    - Bar Chart (if categorical frequencies are needed): Good if we want to see exact counts for each score.
    

```python
# Python Code for Histogram
# Here’s the code using Matplotlib:

import matplotlib.pyplot as plt
import numpy as np

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Summary statistics
mean_score = np.mean(survey_scores)
median_score = np.median(survey_scores)
std_dev = np.std(survey_scores)
min_score, max_score = min(survey_scores), max(survey_scores)

print("Mean:", mean_score)
print("Median:", median_score)
print("Standard Deviation:", std_dev)
print("Min:", min_score, "Max:", max_score)

# Histogram
plt.hist(survey_scores, bins=6, edgecolor="black", alpha=0.7)
plt.title("Customer Satisfaction Survey Distribution")
plt.xlabel("Satisfaction Score (1–10)")
plt.ylabel("Frequency")
plt.axvline(mean_score, color="red", linestyle="dashed", linewidth=1, label=f"Mean = {mean_score:.2f}")
plt.axvline(median_score, color="green", linestyle="dashed", linewidth=1, label=f"Median = {median_score}")
plt.legend()
plt.show()
```

What this does:
- Prints mean, median, standard deviation, min, max.
- Plots a histogram with vertical lines showing the mean & median.
